```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
data = pd.DataFrame(data = wine.data, columns = wine.feature_names)
data["target"] = pd.DataFrame(data = wine.target)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.scatterplot(data = data, x = "alcohol", y="alcalinity_of_ash", hue = "target")
```

```python
sns.catplot(kind = "swarm", data = data, x = "target", y = "alcohol", hue = "target")
```

```python
sns.relplot(kind = "line", data = data, x = "hue", y = "color_intensity", hue = "target")
```

```python
sns.catplot(data = data, x = "target", y = "proline", kind = "box")
```
